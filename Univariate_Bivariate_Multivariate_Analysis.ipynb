# 1. Univariate Analysis

# Univariate Analysis - Barplot (Categorical data)

```python
sns.barplot(df_titanic['Embarked'])
plt.title('Passengers from different location')
plt.show()
```

```python
sns.barplot(df_employee['City'], palette = ['red','green','blue'])
plt.title('Employee from different cities')
plt.show()
```

# Univariate Analysis - Piechart (Categorical data)

```python
data = df_titanic['Survived'].value_counts()

plt.pie(data, labels = data.index, autopct = '%1.2f%%')
plt.title('Survived Distribution')
plt.show()
```

```python
data = df_employee['City'].value_counts()

plt.pie(data, labels = data.index, autopct = '%1.2f%%')
plt.title('Pie chart on employee from different city')
plt.show()
```

# Univariate Analysis - Donutchart (Categorical data)

```python
# Donut chart on employee percentage according to city on employee dataset
data = df_employee['City'].value_counts()

plt.pie(data,
        labels = data.index,
        wedgeprops = {'width':0.5},
        autopct = '%1.2f%%')
plt.title('Pie chart on employee from different city')
plt.show()
```

# Univariate Analysis - Countplot (Categorical data)

# 3. Multivariate Analysis

# Multivariate Analysis - Scatter plot 

```python
sns.scatterplot(data = df_iris, x = 'petal_length', y = 'petal_width', hue = 'species')
plt.title('Scatter plot on petal length and petal width to find species')
plt.show()
```

```python
sns.scatterplot(data = df_titanic, x = 'Age', y = 'Fare', hue = 'Survived')
plt.title('Scatter plot on age, fare and survived')
plt.show()
```

# Multivariate Analysis - Rugplot 

```python
sns.scatterplot(data = df_bill, x = 'total_bill', y = 'tip', hue = 'smoker')
sns.rugplot(data = df_bill, x = 'total_bill', y = 'tip', hue = 'smoker')
plt.title('Rugplot on total bill and tip and somker')
plt.show()
```

# Multivariate Analysis - Pair plot 

```python
sns.pairplot(df_iris, hue = 'species')
plt.show()
```

# Multivariate Analysis - Heatmap

```python
sns.pairplot(df_titanic, hue = 'Survived')
plt.show()
```

```python
sns.heatmap(df_iris.corr(numeric_only = True))
plt.show()
```

# Multivariate Analysis - FacetGrid plot

```python
g = sns.FacetGrid(df_bill,col = 'day',row = 'time')
g.map(sns.scatterplot,'total_bill','tip')
plt.show()
```

```python
g = sns.FacetGrid(df_bill,col = 'day',row = 'time')
g.map_dataframe(sns.histplot,x = 'total_bill')
plt.show()
```

# Multivariate Analysis - 3D Scatter plot

```python
from mpl_toolkits.mplot3d import Axes3D
# Create 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_titanic['Age'],df_titanic['Fare'],df_titanic['Pclass'],c=df_titanic['Survived'])
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Passenger Class')
ax.set_title('3D Scatter Plot of Titanic Dataset')
plt.show()
```

# Multivariate Analysis - PariGrid

```python
g = sns.PairGrid(df_iris)
g.map(sns.scatterplot)
# plt.title('Pairgrid plot in iris dataset')
plt.show()
```

```python
g = sns.PairGrid(df_iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
# plt.title('Pairgrid plot in iris dataset')
plt.show()
```

```python
g = sns.PairGrid(df_iris)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
# g.map_offdiag(sns.scatterplot)
# plt.title('Pairgrid plot in iris dataset')
plt.show()
```

```python
g = sns.PairGrid(df_iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.add_legend()
# g.map_offdiag(sns.scatterplot)
# plt.title('Pairgrid plot in iris dataset')
plt.show()
```
