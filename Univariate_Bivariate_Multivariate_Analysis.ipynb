```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import kagglehub
# kagglehub.dataset_download('<owner>/<dataset-slug>')
```

```python
df_titanic = pd.read_csv('/kaggle/input/datasets/brendan45774/test-file/tested.csv')
df_iris = pd.read_csv('/kaggle/input/datasets/himanshunakrani/iris-dataset/iris.csv')
df_employee = pd.read_csv('/kaggle/input/datasets/tawfikelmetwally/employee-dataset/Employee.csv')
```

# 1. Univariate Analysis

# Univariate Analysis - Barplot (Categorical data)

```python
sns.barplot(df_titanic['Embarked'])
plt.title('Passengers from different location')
plt.show()
```

```python
sns.barplot(df_employee['City'], palette = ['red','green','blue'])
plt.title('Employee from different cities')
plt.show()
```

# Univariate Analysis - Piechart (Categorical data)

```python
data = df_titanic['Survived'].value_counts()

plt.pie(data, labels = data.index, autopct = '%1.2f%%')
plt.title('Survived Distribution')
plt.show()
```

```python
data = df_employee['City'].value_counts()

plt.pie(data, labels = data.index, autopct = '%1.2f%%')
plt.title('Pie chart on employee from different city')
plt.show()
```

# Univariate Analysis - Countplot (Categorical data)

```python
sns.countplot(data = df_titanic, x = 'Sex')
plt.title('Countplot on gender basis')
plt.show()
```

# Univariate Analysis - Histogram (Numerical data)

```python
sns.histplot(data = df_iris, x = 'sepal_length')
plt.title('Flower count of different sepal length')
plt.show()
```

```python
sns.histplot(data = df_iris, x = 'petal_width')
plt.title('Flower count of different petal width')
plt.show()
```

```python
sns.histplot(df_employee['Age'])
plt.title('Age of all employee')
plt.show()
```

# Univariate Analysis - Kdeplot (Numerical data)

```python
sns.kdeplot(data = df_iris, x = 'sepal_length')
plt.title('Flower count of different sepal length')
plt.show()
```

```python
sns.kdeplot(data = df_iris, x = 'petal_width')
plt.title('Flower count of different petal width')
plt.show()
```

```python
sns.kdeplot(df_employee['Age'])
plt.title('Density distribution of age')
plt.show()
```

# Univariate Analysis - Violinplot (Numerical data)

```python
sns.violinplot(data = df_employee['City'])
plt.title('Employee according to city')
plt.show()
```

```python
sns.violinplot(df_titanic['Embarked'],color = 'blue')
plt.title('Passengers from different city in titanic')
plt.show()
```

# Univariate Analysis - Distplot (Numerical data)

```python
sns.distplot(df_employee['Age'])
plt.title('Dist plot on age')
plt.show()
```

```python
sns.distplot(df_iris['sepal_width'])
plt.title('Dist plot on sepal width')
plt.show()
```

# Univariate Analysis - Boxplot (Numerical data)

```python
sns.boxplot(data = df_titanic, x = 'Age')
plt.title('Box plot on age of passengers')
plt.show()
```

```python
sns.boxplot(data = df_titanic, x = 'Fare')
plt.title('Boxplot on Fare of each passenger')
plt.show()
```

# 2. Bivariate Analysis

# Bivariate Analysis - Scatter Plot (Numerical vs Numerical)

```python
sns.scatterplot(data = df_titanic, x = 'Age', y = 'Fare')
plt.title('Scatter plot on age and fare')
plt.show()
```

# Bivariate Analysis - Line Plot (Numerical vs Numerical)

```python
cnt = df_employee['JoiningYear'].value_counts()
sns.lineplot(x = cnt.index, y = cnt.values)
plt.title('Employee count line plot according to joining year')
plt.show()
```

```python
sns.lineplot(data = df_titanic, x = 'Age', y = 'Fare')
```

# Bivariate Analysis - Boxplot (Numerical vs Categorical)

```python
sns.boxplot(data = df_titanic, x = 'Survived', y = 'Age', palette = 'Set2')
plt.title('Boxplot on survived according to age')
plt.show()
```

# Bivariate Analysis - Violinplot (Numerical vs Categorical)

```python
sns.violinplot(data = df_titanic, x = 'Survived', y = 'Pclass', palette = 'Set2')
```

# Bivariate Analysis - Barplot (Numerical vs Categorical)

```python
sns.barplot(data = df_employee, x = 'City', y = 'LeaveOrNot')
```

# Bivariate Analysis - Countplot (Categorical vs Categorical)

```python
sns.countplot(data = df_titanic, x = 'Pclass', hue = 'Survived')
plt.title('Count plot on pclass and survived passenger')
plt.show()
```

# 3. Multivariate Analysis

# Multivariate Analysis - Scatter plot 

```python
sns.scatterplot(data = df_iris, x = 'petal_length', y = 'petal_width', hue = 'species')
plt.title('Scatter plot on petal length and petal width to find species')
plt.show()
```

```python
sns.scatterplot(data = df_titanic, x = 'Age', y = 'Fare', hue = 'Survived')
plt.title('Scatter plot on age, fare and survived')
plt.show()
```

# Multivariate Analysis - Pair plot 

```python
sns.pairplot(df_iris, hue = 'species')
plt.show()
```

# Multivariate Analysis - Heatmap

```python
sns.pairplot(df_titanic, hue = 'Survived')
plt.show()
```

```python
sns.heatmap(df_iris.corr(numeric_only = True))
plt.show()
```
