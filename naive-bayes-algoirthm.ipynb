```python
df_iris = pd.read_csv('/kaggle/input/datasets/organizations/uciml/iris/Iris.csv')
```

```python
df_iris
```

```python
df_iris = df_iris.drop(columns = ['Id'])
```

```python
df_iris
```

```python
x = df_iris.drop(columns = ['Species'])
y = df_iris['Species']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# model creation
model = GaussianNB()
```

```python
# Tran model
model.fit(x_train, y_train)
```

```python
#prediction
y_pred = model.predict(x_test)
```

```python
# accuracy check
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
